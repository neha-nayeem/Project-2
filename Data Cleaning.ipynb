```python
import pandas as pd
```

```python
# Import the crime data
crime_data="Resources/Toronto_Crimes_Occurred_2014_2019.csv"
crime_df = pd.read_csv(crime_data)
crime_df.head()
```

```python
# Select the columns for the projects
df1=crime_df[["offence","occurrenceyear","MCI","Division","Hood_ID","Neighbourhood","Long","Lat"]]
df1.head()
```

```python
# Filter the crimes occured in 2016
df1 = df1.loc[df1["occurrenceyear"]==2016]
df1.reset_index(inplace=True,drop=True)
df1
```

```python
# Import the neighbourhood profile
neighbourhood="Resources/neighbourhood-profiles-2016-csv.csv"
neighbourhood_df=pd.read_csv(neighbourhood)
neighbourhood_df.head()
```

```python
# Filter three demographics - age, income, unemployment.
demographic_df = neighbourhood_df.loc[(neighbourhood_df["_id"] == 1) | (neighbourhood_df["_id"] == 8) | (neighbourhood_df["_id"] == 10) | (neighbourhood_df["_id"] == 11) | (neighbourhood_df["_id"] == 12) | (neighbourhood_df["_id"] == 13) | (neighbourhood_df["_id"] == 14) | (neighbourhood_df["_id"] == 15) | (neighbourhood_df["_id"] == 1030) | (neighbourhood_df["_id"] == 1891)]
demographic_df
```

```python
# Remove columns not required
demographic_df.drop(["_id","Category","Topic","Data Source"],inplace=True,axis=1)
demographic_df
```

```python
# Set up the new index
demographic_df.set_index("Characteristic",inplace=True)
demographic_df
```

```python
# DataFrame transpose
neighbourhood_transposed = demographic_df.T
neighbourhood_transposed
```

```python
# Convert all columns from object to integers
demographic_info=neighbourhood_transposed.replace({',':''},regex=True).apply(pd.to_numeric,1)
demographic_info
```

```python
# Add coloumns of average age and population
demographic_info["Population"] = demographic_info["Children (0-14 years)"]+demographic_info["Youth (15-24 years)"]+demographic_info["Working Age (25-54 years)"]+demographic_info["Pre-retirement (55-64 years)"]+demographic_info["Seniors (65+ years)"]+demographic_info["Older Seniors (85+ years)"]
demographic_info["Average age"] = (7*demographic_info["Children (0-14 years)"]
                                   +19.5*demographic_info["Youth (15-24 years)"]
                                   +39.5*demographic_info["Working Age (25-54 years)"]
                                   +59.5*demographic_info["Pre-retirement (55-64 years)"]
                                   +75*demographic_info["Seniors (65+ years)"]
                                   +85*demographic_info["Older Seniors (85+ years)"])/demographic_info["Population"]
demographic_info
```

```python
# Reset index
demographic_info.reset_index(inplace=True)
demographic_info.rename(columns={"index":"Neighbourhood Name","Neighbourhood Number":"Hood_ID"},inplace=True)
demographic_info
```

```python
# Merge the demographic data with the crime data
merged_df = pd.merge(df1,demographic_info,on="Hood_ID",how="inner")
merged_df.drop(["Neighbourhood"],inplace=True,axis=1)
merged_df
```

```python
grouped_df=merged_df.groupby(["Neighbourhood Name","MCI"]).count()
crime_df=grouped_df[["offence"]]
crime_df.rename(columns={"offence":"number_of_crime"},inplace=True)
crime_df
```

```python
# Export cleaned data to csv
df1.to_csv("Data-Output/crime_data_2016.csv",index=False)
demographic_info.to_csv("Data-Output/neighbourhood_profile_data.csv",index=False)
merged_df.to_csv("Data-Output/neighbourhood_crime_2016.csv",index=False)
crime_df.to_csv("Data-Output/number_of_crime_2016.csv",index=True)
```

```python
new_data="Data-Output/number_of_crime_2016.csv"
new_df = pd.read_csv(new_data)
new_df
```
